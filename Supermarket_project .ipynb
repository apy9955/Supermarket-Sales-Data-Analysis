```python
sns.countplot(x='Gender', data=df , palette='Set2')
plt.title("Customer Gender Distribution")
plt.show()
print(df['Gender'].value_counts())
```

```python
sns.histplot(df["Rating"], kde=True, bins=15, color="skyblue")
plt.title("Customer Rating Distribution")
plt.show()
print("skewness:", skew(df["Rating"]))
```

```python
branch_sales = df.groupby('Branch')['Total'].sum().sort_values(ascending=False)
sns.barplot(x=branch_sales.index, y=branch_sales.values, palette="muted")
plt.title("Aggregate Sales by Branch")
plt.show()
print(branch_sales)
```

```python
sns.countplot(x='Payment', data=df, palette='coolwarm')
plt.title("Payment Method Distribution")
plt.show()
print(df['Payment'].value_counts())
```

```python
sns.countplot(x='Payment', hue='Branch', data=df, palette='Set1')
plt.title("Payment by Branch")
plt.show()
```

```python
sns.scatterplot(x='gross income', y='Rating', data=df)
plt.title("Gross Income vs Rating")
plt.show()
print(df[['gross income','Rating']].corr())
```

```python
sns.boxplot(x='Branch', y='Rating', data=df, palette='Set3')
plt.title("Ratings per Branch")
plt.show()
```

```python
profit_branch = df.groupby('Branch')['gross income'].sum().sort_values(ascending=False)
sns.barplot(x=profit_branch.index, y=profit_branch.values, palette="Blues")
plt.title("Gross Income by Branch")
plt.show()
print(profit_branch)
```

```python
sns.barplot(x='Gender', y='gross income', data=df, estimator=sum, palette='pastel')
plt.title("Gross Income by Gender")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
time_trend = df.groupby('Date')['gross income'].sum()
time_trend.plot(figsize=(12,5))
plt.title("Gross Income Over Time")
plt.show()
```

```python
product_income = df.groupby('Product line')['gross income'].sum().sort_values(ascending=False)
sns.barplot(y=product_income.index, x=product_income.values, palette="viridis")
plt.title("Income by Product Line")
plt.show()
print(product_income)
```

```python
sns.barplot(x='Gender', y='Total', hue='Product line', data=df, estimator=sum)
plt.title("Spending by Gender and Product Line")
plt.show()
```

```python
print("Total Products Bought:", df['Quantity'].sum())
```

```python
df['Day'] = df['Date'].dt.day_name()
day_sales = df.groupby('Day')['Total'].sum().sort_values(ascending=False)
sns.barplot(x=day_sales.index, y=day_sales.values, palette="mako")
plt.title("Sales by Day of Week")
plt.show()
print(day_sales)
```

```python
df['Hour'] = pd.to_datetime(df['Time']).dt.hour
hourly_sales = df.groupby('Hour')['Total'].sum()
hourly_sales.plot(kind='bar', figsize=(12,5), color="coral")
plt.title("Sales by Hour of the Day")
plt.show()
print(hourly_sales)
```
